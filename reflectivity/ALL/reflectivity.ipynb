# Initialize packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Import the data

```python
# Declare the data filename

ref_5x5 = "reflectivity_5x5.csv"
ref_1Mo = "reflectivity_1Mo.csv"
ref_1S = "reflectivity_1S.csv"
ref_2Mo = "reflectivity_2Mo.csv"
ref_2S = "reflectivity_2S.csv"
ref_3Mo = "reflectivity_3Mo.csv"
ref_3S = "reflectivity_3S.csv"
ref_4Mo = "reflectivity_4Mo.csv"
ref_4S = "reflectivity_4S.csv"
ref_4Mo_2 = "reflectivity_4Mo_2.csv"
ref_4S_2 = "reflectivity_4S_2.csv"
ref_5Mo = "reflectivity_5Mo.csv"
ref_5S = "reflectivity_5S.csv"

# Define the header of the dataset

ColumnName = ['Energy_eV', 'Reflectivity']

```

Read the data  datafile variable of a string of the file

```python
#Read the data

reflect_5x5 = pd.read_csv(ref_5x5, skiprows=1, names=ColumnName)
reflect_1Mo = pd.read_csv(ref_1Mo, skiprows=1, names=ColumnName)
reflect_1S = pd.read_csv(ref_1S, skiprows=1, names=ColumnName)
reflect_2Mo = pd.read_csv(ref_2Mo, skiprows=1, names=ColumnName)
reflect_2S = pd.read_csv(ref_2S, skiprows=1, names=ColumnName)
reflect_3Mo = pd.read_csv(ref_3Mo, skiprows=1, names=ColumnName)
reflect_3S = pd.read_csv(ref_3S, skiprows=1, names=ColumnName)
reflect_4Mo = pd.read_csv(ref_4Mo, skiprows=1, names=ColumnName)
reflect_4S = pd.read_csv(ref_4S, skiprows=1, names=ColumnName)
reflect_4Mo_2 = pd.read_csv(ref_4Mo_2, skiprows=1, names=ColumnName)
reflect_4S_2 = pd.read_csv(ref_4S_2, skiprows=1, names=ColumnName)
reflect_5Mo = pd.read_csv(ref_5Mo, skiprows=1, names=ColumnName)
reflect_5S = pd.read_csv(ref_5S, skiprows=1, names=ColumnName)
#Show first five datapoints

reflect_5x5.head()
```

```python
#Declare the figure and size

pic_ref_5x5 = plt.figure(figsize=(8,5))

#Add a data to the plot (using column name)
##"DataSetName.ColumnName"
##Parameter references: https://matplotlib
plt.plot(reflect_5x5.Energy_eV, reflect_5x5.Reflectivity, color='k', ls='-')

#Plot details
#plt.title("Density of states plot") # we don't use this in DOS plots
plt.ylabel('Reflectivity', size=14)
plt.xlabel('Energy (eV)', size=14)

#Plot x and y range

plt.xlim(0.04, 5)
plt.ylim(0, 0.6)

#activates legends
#plt.legend(loc='upper right', prop={'size': 14})

#Define filename

fname = "reflectivity_5x5.png"

#Save the plot

pic_ref_5x5.savefig(fname, dpi=500, bbox_inches='tight')
```
