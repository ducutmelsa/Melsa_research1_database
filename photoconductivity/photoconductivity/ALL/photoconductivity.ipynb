# Initialize packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Import the data

```python
# Declare the data filename

ph_5x5 = "photoconductivity_5x5.csv"
ph_1Mo = "photoconductivity_1Mo.csv"
ph_1S = "photoconductivity_1S.csv"
ph_2Mo = "photoconductivity_2Mo.csv"
ph_2S = "photoconductivity_2S.csv"
ph_3Mo = "photoconductivity_3Mo.csv"
ph_3S = "photoconductivity_3S.csv"
ph_4Mo = "photoconductivity_4Mo.csv"
ph_4S = "photoconductivity_4S.csv"
ph_4Mo_2 = "photoconductivity_4Mo_2.csv"
ph_4S_2 = "photoconductivity_4S_2.csv"
ph_5Mo = "photoconductivity_5Mo.csv"
ph_5S = "photoconductivity_5S.csv"

# Define the header of the dataset

ColumnName = ['Energy_eV', 'Photoconductivity']

```

Read the data  datafile variable of a string of the file

```python
#Read the data

photo_5x5 = pd.read_csv(ph_5x5, skiprows=1, names=ColumnName)
photo_1Mo = pd.read_csv(ph_1Mo, skiprows=1, names=ColumnName)
photo_1S = pd.read_csv(ph_1S, skiprows=1, names=ColumnName)
photo_2Mo = pd.read_csv(ph_2Mo, skiprows=1, names=ColumnName)
photo_2S = pd.read_csv(ph_2S, skiprows=1, names=ColumnName)
photo_3Mo = pd.read_csv(ph_3Mo, skiprows=1, names=ColumnName)
photo_3S = pd.read_csv(ph_3S, skiprows=1, names=ColumnName)
photo_4Mo = pd.read_csv(ph_4Mo, skiprows=1, names=ColumnName)
photo_4S = pd.read_csv(ph_4S, skiprows=1, names=ColumnName)
photo_4Mo_2 = pd.read_csv(ph_4Mo_2, skiprows=1, names=ColumnName)
photo_4S_2 = pd.read_csv(ph_4S_2, skiprows=1, names=ColumnName)
photo_5Mo = pd.read_csv(ph_5Mo, skiprows=1, names=ColumnName)
photo_5S = pd.read_csv(ph_5S, skiprows=1, names=ColumnName)
#Show first five datapoints

photo_5x5.head()
```

```python
#Declare the figure and size

pic_ph_5x5 = plt.figure(figsize=(8,5))

#Add a data to the plot (using column name)
##"DataSetName.ColumnName"
##Parameter references: https://matplotlib
plt.plot(photo_5x5.Energy_eV, photo_5x5.Photoconductivity, color='k', ls='-')

#Plot details
#plt.title("Density of states plot") # we don't use this in DOS plots
plt.ylabel('Photoconductivity (1/fs)', size=14)
plt.xlabel('Energy (eV)', size=14)

#Plot x and y range

plt.xlim(0.04, 5)
plt.ylim(0, 1.0)

#activates legends
#plt.legend(loc='upper right', prop={'size': 14})

#Define filename

fname = "Photoconductivity_5x5.png"

#Save the plot

pic_ph_5x5.savefig(fname, dpi=500, bbox_inches='tight')
```
