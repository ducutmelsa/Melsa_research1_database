# Initialize packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Import the data

```python
# Declare the data filename

ab_5x5 = "absorption_coefficient_5x5.csv"
ab_1Mo = "absorption_coefficient_1Mo.csv"
ab_1S = "absorption_coefficient_1S.csv"
ab_2Mo = "absorption_coefficient_2Mo.csv"
ab_2S = "absorption_coefficient_2S.csv"
ab_3Mo = "absorption_coefficient_3Mo.csv"
ab_3S = "absorption_coefficient_3S.csv"
ab_4Mo = "absorption_coefficient_4Mo.csv"
ab_4S = "absorption_coefficient_4S.csv"
ab_4Mo_2 = "absorption_coefficient_4Mo_2.csv"
ab_4S_2 = "absorption_coefficient_4S_2.csv"
ab_5Mo = "absorption_coefficient_5Mo.csv"
ab_5S = "absorption_coefficient_5S.csv"

# Define the header of the dataset

ColumnName = ['Energy_eV', 'Absorption_Coefficient']

```

Read the data  datafile variable of a string of the file

```python
#Read the data

absorpt_5x5 = pd.read_csv(ab_5x5, skiprows=1, names=ColumnName)
absorpt_1Mo = pd.read_csv(ab_1Mo, skiprows=1, names=ColumnName)
absorpt_1S = pd.read_csv(ab_1S, skiprows=1, names=ColumnName)
absorpt_2Mo = pd.read_csv(ab_2Mo, skiprows=1, names=ColumnName)
absorpt_2S = pd.read_csv(ab_2S, skiprows=1, names=ColumnName)
absorpt_3Mo = pd.read_csv(ab_3Mo, skiprows=1, names=ColumnName)
absorpt_3S = pd.read_csv(ab_3S, skiprows=1, names=ColumnName)
absorpt_4Mo = pd.read_csv(ab_4Mo, skiprows=1, names=ColumnName)
absorpt_4S = pd.read_csv(ab_4S, skiprows=1, names=ColumnName)
absorpt_4Mo_2 = pd.read_csv(ab_4Mo_2, skiprows=1, names=ColumnName)
absorpt_4S_2 = pd.read_csv(ab_4S_2, skiprows=1, names=ColumnName)
absorpt_5Mo = pd.read_csv(ab_5Mo, skiprows=1, names=ColumnName)
absorpt_5S = pd.read_csv(ab_5S, skiprows=1, names=ColumnName)
#Show first five datapoints

absorpt_5x5.head()
```

```python
#Declare the figure and size

pic_ab_5x5 = plt.figure(figsize=(8,5))

#Add a data to the plot (using column name)
##"DataSetName.ColumnName"
##Parameter references: https://matplotlib
plt.plot(absorpt_5x5.Energy_eV, absorpt_5x5.Absorption_Coefficient, color='k', ls='-')

#Plot details
#plt.title("Density of states plot") # we don't use this in DOS plots
plt.ylabel('Absorption Coefficient (1/cm)', size=14)
plt.xlabel('Energy (eV)', size=14)

#Plot x and y range

plt.xlim(0.04, 5)
plt.ylim(0, 140000)

#activates legends
#plt.legend(loc='upper right', prop={'size': 14})

#Define filename

fname = "abscorp_5x5.png"

#Save the plot

pic_ab_5x5.savefig(fname, dpi=500, bbox_inches='tight')
```
