# Initialize packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Import the data

```python
# Declare the data filename

epsr_5x5 = "epsr_5x5.dat"
epsr_1Mo = "epsr_1Mo.dat"
epsr_1S = "epsr_1S.dat"
epsr_2Mo = "epsr_2Mo.dat"
epsr_2S = "epsr_2S.dat"
epsr_3Mo = "epsr_3Mo.dat"
epsr_3S = "epsr_3S.dat"
epsr_4Mo = "epsr_4Mo.dat"
epsr_4Mo_2 = "epsr_4Mo_2.dat"
epsr_4S = "epsr_4S.dat"
epsr_4S_2 = "epsr_4S_2.dat"
epsr_5Mo = "epsr_5Mo.dat"
epsr_5S = "epsr_5S.dat"

# Define the header of the dataset

ColumnName = ['Energy_eV', 'epsi_x', 'epsi_y', 'epsi_z']
```

Read the data  datafile variable of a string of the file

```python
#Read the data

re_5x5 = pd.read_csv(epsr_5x5, skiprows=3, delim_whitespace=True, names=ColumnName)
re_1Mo = pd.read_csv(epsr_1Mo, skiprows=3, delim_whitespace=True, names=ColumnName)
re_1S = pd.read_csv(epsr_1S, skiprows=3, delim_whitespace=True, names=ColumnName)
re_2Mo = pd.read_csv(epsr_2Mo, skiprows=3, delim_whitespace=True, names=ColumnName)
re_2S = pd.read_csv(epsr_2S, skiprows=3, delim_whitespace=True, names=ColumnName)
re_3Mo = pd.read_csv(epsr_3Mo, skiprows=3, delim_whitespace=True, names=ColumnName)
re_3S = pd.read_csv(epsr_3S, skiprows=3, delim_whitespace=True, names=ColumnName)
re_4Mo = pd.read_csv(epsr_4Mo, skiprows=3, delim_whitespace=True, names=ColumnName)
re_4Mo_2 = pd.read_csv(epsr_4Mo_2, skiprows=3, delim_whitespace=True, names=ColumnName)
re_4S = pd.read_csv(epsr_4S, skiprows=3, delim_whitespace=True, names=ColumnName)
re_4S_2 = pd.read_csv(epsr_4S_2, skiprows=3, delim_whitespace=True, names=ColumnName)
re_5Mo = pd.read_csv(epsr_5Mo, skiprows=3, delim_whitespace=True, names=ColumnName)
re_5S = pd.read_csv(epsr_5S, skiprows=3, delim_whitespace=True, names=ColumnName)

#Read Head
re_5x5.head()
```

```python
#Declare the figure and size

fig_5x5 = plt.figure(figsize=(8,5))

#Add a data to the plot (using column name)
##"DataSetName.ColumnName"
##Parameter references: https://matplotlib
plt.plot(re_5x5.Energy_eV, re_5x5.epsi_x, color='k', ls='-', label = '${\epsilon}\mathrm{_2}$')

#Plot details
#plt.title("Density of states plot") # we don't use this in DOS plots
plt.ylabel('Real Part of the Dielectric Function', size=14)
plt.xlabel('Energy (eV)', size=14)

#Plot x and y range

plt.xlim(0.04, 5)
plt.ylim(0,14)
#plt.xticks(ticks = (0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0))
#activates legends
#plt.legend(loc='upper right', prop={'size': 14})

#Define filename

fname = "5x5_epsr.png"

#Save the plot

fig_5x5.savefig(fname, dpi=500, bbox_inches='tight')
```
