# Initialize packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Import the data

```python
# Declare the data filename

epsi_5x5 = "epsi_5x5.dat"
epsi_1Mo = "epsi_1Mo.dat"
epsi_1S = "epsi_1S.dat"
epsi_2Mo = "epsi_2Mo.dat"
epsi_2S = "epsi_2S.dat"
epsi_3Mo = "epsi_3Mo.dat"
epsi_3S = "epsi_3S.dat"
epsi_4Mo = "epsi_4Mo.dat"
epsi_4Mo_2 = "epsi_4Mo_2.dat"
epsi_4S = "epsi_4S.dat"
epsi_4S_2 = "epsi_4S_2.dat"
epsi_5Mo = "epsi_5Mo.dat"
epsi_5S = "epsi_5S.dat"

# Define the header of the dataset

ColumnName = ['Energy_eV', 'epsi_x', 'epsi_y', 'epsi_z']
```

Read the data  datafile variable of a string of the file

```python
#Read the data

im_5x5 = pd.read_csv(epsi_5x5, skiprows=2, delim_whitespace=True, names=ColumnName)
im_1Mo = pd.read_csv(epsi_1Mo, skiprows=2, delim_whitespace=True, names=ColumnName)
im_1S = pd.read_csv(epsi_1S, skiprows=2, delim_whitespace=True, names=ColumnName)
im_2Mo = pd.read_csv(epsi_2Mo, skiprows=2, delim_whitespace=True, names=ColumnName)
im_2S = pd.read_csv(epsi_2S, skiprows=2, delim_whitespace=True, names=ColumnName)
im_3Mo = pd.read_csv(epsi_3Mo, skiprows=2, delim_whitespace=True, names=ColumnName)
im_3S = pd.read_csv(epsi_3S, skiprows=2, delim_whitespace=True, names=ColumnName)
im_4Mo = pd.read_csv(epsi_4Mo, skiprows=2, delim_whitespace=True, names=ColumnName)
im_4Mo_2 = pd.read_csv(epsi_4Mo_2, skiprows=2, delim_whitespace=True, names=ColumnName)
im_4S = pd.read_csv(epsi_4S, skiprows=2, delim_whitespace=True, names=ColumnName)
im_4S_2 = pd.read_csv(epsi_4S_2, skiprows=2, delim_whitespace=True, names=ColumnName)
im_5Mo = pd.read_csv(epsi_5Mo, skiprows=2, delim_whitespace=True, names=ColumnName)
im_5S = pd.read_csv(epsi_5S, skiprows=2, delim_whitespace=True, names=ColumnName)

#Read Head
im_5x5.head()
```

```python
#Declare the figure and size

fig_5x5 = plt.figure(figsize=(8,5))

#Add a data to the plot (using column name)
##"DataSetName.ColumnName"
##Parameter references: https://matplotlib
plt.plot(im_5x5.Energy_eV, im_5x5.epsi_x, color='k', ls='-', label = '${\epsilon}\mathrm{_2}$')

#Plot details
#plt.title("Density of states plot") # we don't use this in DOS plots
plt.ylabel('Imaginary Part of the Dielectric Function', size=14)
plt.xlabel('Energy (eV)', size=14)

#Plot x and y range

plt.xlim(0.04, 5)
plt.ylim(0,16)
#plt.xticks(ticks = (0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0))
#activates legends
#plt.legend(loc='upper right', prop={'size': 14})

#Define filename

fname = "5x5_epsi.png"

#Save the plot

fig_5x5.savefig(fname, dpi=500, bbox_inches='tight')
```
